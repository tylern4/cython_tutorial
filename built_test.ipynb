```python
%matplotlib inline
import cpp.build.wave_propogation as wave_propogation
import time
import matplotlib.pylab as plt
import numpy as np
```

```python
num_steps = 500
scale = 50
damping= 0.5
initial_P = 250

start = time.time()
pressure = wave_propogation.wave_propogation_cpp(num_steps, scale, damping, initial_P)
end = time.time()

print(f'{end-start} S ===> {num_steps/(end-start)} Hz')
plt.figure(figsize=(10,10))
plt.imshow(pressure,cmap='viridis_r',interpolation='lanczos')
plt.colorbar()
```
